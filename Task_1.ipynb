```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/drive/MyDrive/Assignment- Respirer/cons_data_all_res.xlsx')
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df.set_index('Time', inplace=True) # Setting up the main index as the time of the dataframe
```

```python
seasonal = df.resample('Q').mean()  # Quarterly analysis
monthly = df.resample('M').mean()    # Monthly analysis
daily = df.resample('D').mean()      # Daily aalysis
hourly = df.resample('H').mean()     # Hourly analysis
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
seasonal['r1'].plot(title='Seasonal Analysis', color = 'green') # Can change the parameter as per convenience.
plt.subplot(2, 2, 2)
monthly['r1'].plot(title='Monthly Analysis')
plt.subplot(2, 2, 3)
daily['r1'].plot(title='Daily Analyss',color = 'red')
plt.subplot(2, 2, 4)
hourly['r1'].plot(title='Hourly Analysis',color = 'purple')
plt.tight_layout()
plt.show()
```
