```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/drive/MyDrive/Assignment- Respirer/cons_data_all_res.xlsx')
```

The most important reason that I am choosing K-Means clustering here is that it is very effeicent to use and it works on large datasets too. Moreover, I've tried different values of clusters but chose as 3 here.

```python
r_columns = ['r{}'.format(i) for i in range(1, 13)] # its a method to choose columns easily
df1 = df[r_columns].copy()

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3)  # You can change the number of clusters as needed
cl = kmeans.fit_predict(df1)
df1['Cluster'] = cl
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df1['r3'], df1['r4'], c=cl)
plt.xlabel('r3')
plt.ylabel('r4')
plt.title('Clustered Analysis r3 vs r4') # Chose any two parameter to visualize them
plt.show()
```
